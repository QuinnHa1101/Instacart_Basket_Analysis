## 03. Visualization for key questions

### 3.1. The busiest days of the week and hours of the day

```python
# Create bar chart for orders_day_week
bar_dow = df_high['orders_day_of_week'].value_counts().sort_index().plot.bar(color = '#77aa63', rot = 0)
plt.title('Busiest days of the week')
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.show()
```

```python
# Exporting bar chart
bar_dow.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'final_orders_dow.png'))
```

```python
# Create a histogram for order hour of day
hist_order_hour = df_high['order_hour_of_day'].plot.hist(bins = 24, grid = False, color = '#77aa63')
plt.title('Frequency of hour of order')
plt.xlabel('Hour')
plt.ylabel('Frequency')
plt.xticks(np.arange(0,24,2))
plt.show()
```

```python
# Exporting bar chart
hist_order_hour.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'final_hist_order_hour.png'))
```

### 3.2. Expenditure by hour of day

```python
# Creat line chart
line_hour = sns.lineplot(data = df_high, x = 'order_hour_of_day',y = 'prices')
plt.title('Difference in expenditure during the day')
plt.xlabel('Hour of day')
plt.ylabel('Prices')
plt.show()
```

```python
# Exporting chart
line_hour.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'final_line_expenditure_hour.png'))
```

### 3.3. Simpler price range groupings

```python
price_groups = df_high['price_range_loc'].value_counts(dropna = False)
```

```python
price_groups
```

```python
bar_price_group = price_groups.plot.bar(color = '#77aa63', rot = 0)
plt.title('Number of each Price tag')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.show()
```

```python
# Exporting chart
bar_price_group.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'final_bar_price_group.png'))
```

```python
pie_price_group = price_groups.plot.pie(colors = ['#a1464b','#78964b','#d6cc00'], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Percentage of each Price tag')
plt.ylabel(' ')
plt.show()
```

```python
# Exporting chart
pie_price_group.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'final_pie_price_group.png'))
```

### 3.4. Popular products & departments

```python
# Create bar chart departments
bar_department = df_high['department_id'].value_counts(dropna = False).plot.bar(color = '#77aa63', rot = 0)
plt.title('Popularity of each department')
plt.ylabel('Frequency')
plt.xlabel('Department ID')
plt.show()
```

```python
# Exporting chart
bar_department.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'popular_department.png'))
```

```python
# Create bar chart top 10 products
bar_product = df_high['product_name'].value_counts(dropna = False).head(10).sort_values(ascending = True).plot(kind = 'barh', color = '#77aa63',y='product_name')
fig = plt.gcf()
fig.set_size_inches(15, 8)
plt.title('Popularity of the top 10 products')
plt.ylabel(' ')
plt.xlabel('Frequency')
plt.show()
```

```python
# Exporting chart
bar_product.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'popular_product.png'))
```

### 3.1 Create a bar chart

```python
# Create bar chart for orders_day_week
bar = ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['purple','red','pink','orange','yellow','blue','green'])
```

```python
# Exporting bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

### 3.2 Create Histograms and Scatterplots

```python
# Create histogram chart for prices
hist = ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
hist_2 = ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# Create scatterplots
plot = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

### 3.3 Create a line graph

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```
